```python
import pandas as pd
import matplotlib.pyplot as plt
from pyjack21.game import BlackJackTable
```

```python
player_count = 4
payroll = [10000, 10000, 10000, 10000]
table = BlackJackTable(hands=2000, decks=6, player_count=player_count, payroll=payroll)
```

```python
table.run()
```

```python
table.df
```

```python
blackjack_data = table.df
for i in range(player_count):
    plt.plot(blackjack_data[f'payroll_{i}'])
```
